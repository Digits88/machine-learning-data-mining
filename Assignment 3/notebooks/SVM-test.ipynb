```python
import pandas

data = pandas.read_csv("../data/owls.csv")
```

```python
from sklearn.svm import LinearSVC
import numpy as np

total = len(data)
split = int(np.floor(total * 0.6))
samples = np.arange(total)
np.random.shuffle(samples)

training = samples[:split]
testing = samples[split:]

svm = LinearSVC()

reduced =  svm.fit_transform(data.iloc[training, :-1], data.iloc[training, -1])

for i, x in enumerate(reduced):
    print "%r -> %r" % (data.iloc[i, :-1].values, x)

predictions = svm.predict(data.iloc[testing, :-1])
actuals = data.iloc[testing, -1]

for i, prediction in enumerate(predictions):
    actual = actuals.iloc[i]
    
    print "%s \t==\t%s \t(%r)" % (actual, prediction, actual == prediction)
    
```
